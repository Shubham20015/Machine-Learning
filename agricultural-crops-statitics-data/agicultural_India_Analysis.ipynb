Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (12, 9)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
df = data_file[["Crop","Cost of Cultivation (`/Hectare) A2+FL","Cost of Cultivation (`/Hectare) C2"]].copy()
```

```python
df = df.groupby('Crop')[['Cost of Cultivation (`/Hectare) A2+FL','Cost of Cultivation (`/Hectare) C2']].mean().reset_index()
```

```python
fig = df.plot(x="Crop",y=["Cost of Cultivation (`/Hectare) A2+FL","Cost of Cultivation (`/Hectare) C2"],kind="bar",
              title='Cost of Cultivation A2+FL vs C2')
```

This bar graph shows how there is significant difference between Cost of cultivation by (A2+FL vs C2).

The Commission for Agricultural Costs & Prices (CACP) in the Ministry of Agriculture projects three kinds of production cost for every crop, both at state and all-India average levels. ‘A2’ covers all paid-out costs directly incurred by the farmer — in cash and kind — on seeds, fertilisers, pesticides, hired labour, leased-in land, fuel, irrigation, etc. ‘A2+FL’ includes A2 plus an imputed value of unpaid family labour. ‘C2’ is a more comprehensive cost that factors in rentals and interest forgone on owned land and fixed capital assets, on top of A2+FL.

## Asking and Answering Questions

We've already gained several insights about the crops and the indian agriculture community by exploring individual columns of the dataset. Let's ask some specific questions and try to answer them using data frame operations and visualizations.



#### Q1: Which State produces what types of crops ?

```python
def getCropList(crops):
    crops = list(crops)
    list_of_crops = ''
    for crop_name in crops:
        list_of_crops += (crop_name.capitalize() + ', ')
    return list_of_crops   
```

```python
data = data_file['State'].unique()
for state in data:
    crops_array = data_file[data_file.State == state]['Crop']
    crops_list = getCropList(crops_array)
    print("\n '\033[1m {}' are producers of {} crops.".format(state,crops_list))
```

#### Q2: Which states producing highest yield  given crop with least cost of production

#### Q4: What is advisable duration for growing most of the crops ?

```python
# USe to supress warnings
import warnings
warnings.filterwarnings('ignore')

df_4 = zone_file['Season/ duration in days'].str.replace(r'[^\d.]+', '')
df_4 = df_4.replace("", float("Nan")).dropna()
df_4 = df_4.astype('int64')
s = df_4.to_frame(name='Duration')
first = len(s[(s['Duration'] >= 80) & (s['Duration'] <= 100)])
second = len(s[(s['Duration'] > 100) & (s['Duration'] <= 135)])
third = len(s[(s['Duration'] > 135) & (s['Duration'] <= 160)])
fourth = len(s[(s['Duration'] > 160) & (s['Duration'] <= 280)])
five = len(s[(s['Duration'] > 280) & (s['Duration'] <= 360)])
cul_data = {'Duration':['2-3 Months','3-4.5 Months','4.5-7 Months','7-9 Months','Almost a year'],
            'No. of people advice':[first,second,third,fourth,five]}
df_4 = pd.DataFrame(cul_data)
df_4.plot(x='Duration', y='No. of people advice', kind='bar', color='orange',legend=None)
plt.ylabel('No. of people advice')
plt.show()
```

## Inferences and Conclusions

We've drawn many inferences from the survey. Here's a summary of a few of them:

- Based on the survey demographics, we can infer that the survey is somewhat representative of the overall agricultural condition in india. However, it has fewer responses for latest years but maybe it will be enough to reveal how's condition of indian agricultural crops and lands.

- The data shows that there is a variety of crops grown in india. Although yield per hectare is not up to mark according to population of india and also consider the fact that india occupies a leading position in global trade of agricultural products.

- Although there are so many which produce varities of crops but Karnataka,Tamil Nadu,Andhra Pradesh,Maharashtra,Uttar Pradesh are lead producers of crops in india.

- As going through years by years, we see significant increase in percentages of production of different crops.But this is also insufficient due to growing demand of food grains.

- Wheat & Paddy are those crops for which there is high numbers of varities present, closely followed by Pearl Millet & Maize.

- Sugarcane has highest yield (Quintal/Hectare) among all crops with the least cost of production per hectare .

- This survey also reveal an interesting fact that indian species require less area to grow and give highest production/yield.

- Experts/Farmers suggests that duration of most of crops are between 3 to 5 months.
